```python
from pythainlp import word_tokenize
```

```python
text = "ก็จะรู้ความชั่วร้ายที่ทำไว้         และคงจะไม่ยอมให้ทำนาบนหลังคน "
```

```python
print(word_tokenize(text, engine="longest"))
print(word_tokenize(text, engine="mm"))
print(word_tokenize(text, engine="newmm"))
print(word_tokenize(text, engine="newmm-safe"))
print(word_tokenize(text, engine="icu"))
```

```python
print(word_tokenize(text, engine="deepcut"))
print(word_tokenize(text, engine="attacut"))
```

```python
%%timeit
print(word_tokenize(text, engine="longest"))
```

```python
import requests
url_lextop = "https://api.aiforthai.in.th/lextoplus"
url_tlexp = "https://api.aiforthai.in.th/tlexplus"
headers = {"Apikey": ""}

def ai4thai_word_tokenize(url, header, text):
  params = {"text": text, "norm": 1}
  res = requests.get(url, params=params, headers=headers).json()
  return res['tokens']
```

```python
text = "พนักงานโรงแรม Hilton ให้บริิการต้อนรับดีมาก"
```

```python
print(ai4thai_word_tokenize(url_lextop, headers, text))
```

```python
print(ai4thai_word_tokenize(url_lextop, headers, text))
```
